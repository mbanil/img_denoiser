```python
from skimage.util import random_noise
from skimage.restoration import denoise_nl_means, estimate_sigma
import numpy as np
from matplotlib import pyplot as plt
import tifffile
import os
from skimage.metrics import structural_similarity as ssim

from denoisemain import denoiser
```

```python
# Open the image file
img = tifffile.imread("C:/Users/Anil Kumar/Downloads/Seimens stars thick/average_corrected_img.tif")
print(img.shape)
img = (img - np.min(img))/(np.max(img) - np.min(img))

# Display the image
plt.figure(figsize=(7,7))
plt.imshow(img, cmap = 'gray')
```

```python
patch_size = 40
denoised = denoiser(img, clustering_factor=2.7,max_patches_per_class=200, termination_number=3, patch_size=patch_size)
```

```python
tifffile.imsave("C:/Users/Anil Kumar/Downloads/Seimens stars thick/average_corrected_img_denoised.tif", denoised)
```
